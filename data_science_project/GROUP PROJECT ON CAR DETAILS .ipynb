```python
#to find how many different car companies and each company repeating no of times
n = car_df["name"].count()
car_company = car_df["name"]
list1 = []
for i in range(n):
    car_name = car_company[i].split(" ")
    list1.append(car_name[0])
car_company_df = pd.DataFrame(list1,columns = ["name"])
car_company_df.name.value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
sns.countplot(data=car_company_df,x="name",palette="CMRmap")
plt.xticks(rotation=90)
plt.xlabel("Car name",fontsize=15,color="black")
plt.ylabel("Car count",fontsize=15,color="black")
plt.title("Cars count",fontsize=25)
plt.show()
```

```python
car_df.year.value_counts()
```

```python
#count of number of cars in each year......
plt.figure(figsize=(10,10))
sns.countplot(data=car_df,x="year",palette="icefire")
plt.xticks(rotation=90)
plt.xlabel("YEAR",fontsize=15,color="RED")
plt.ylabel("COUNT",fontsize=15,color="RED")
plt.title("YEAR COUNT",fontsize=25,color="RED")
plt.show()
```

```python
#displaying the company names with percentages in piechart among top 25 companies.....
labels = car_company_df["name"][:25].value_counts().index
sizes = car_company_df["name"][:25].value_counts()
plt.figure(figsize = (10,10))
plt.pie(sizes, labels=labels, rotatelabels=False, autopct='%1.1f%%',shadow=True, startangle=45,textprops={'fontsize':20})
plt.title('Car name',color = 'red',fontsize = 30)
plt.show()
```

```python
#displaying the piechart with top 25 datasets with its years......
labels = car_df["year"][:25].value_counts().index
sizes = car_df["year"][:25].value_counts()
plt.figure(figsize = (10,10))
plt.pie(sizes, labels=labels, rotatelabels=False, autopct='%1.1f%%',shadow=True, startangle=45,textprops={'fontsize':20})
plt.title('Year',color = 'red',fontsize = 30)
plt.show()
```

```python
#individual seller count of each of the sector....
plt.figure(figsize=(8,8))
sns.countplot(data=car_df,x="seller_type",palette="icefire")
plt.xticks(rotation=90)
plt.xlabel("Seller type",fontsize=15,color="black")
plt.ylabel("Count",fontsize=15,color="black")
plt.title("Seller Count",fontsize=25,color="black")
plt.show()
```

```python
#transmission of both mannual and automatic in each of the year.....
plt.figure(figsize=(8,8))
sns.histplot(data=car_df, x="year", hue="transmission")
plt.xticks(rotation=45)
plt.show()
```

```python
#box plot of each of the car company....
f, ax = plt.subplots(figsize=(20,8))
sns.boxplot(x=car_company_df["name"].values, y = car_df["km_driven"].values,palette="twilight",ax=ax)
plt.xticks(rotation=90)
plt.show()
```

```python
#ploting the different fuels of the cars in the given dataset......
plt.figure(figsize=(10,5))
sns.lineplot(data=car_df,x='year',y='selling_price',hue='fuel')
```

```python
sns.pairplot(car_df)
plt.show()
```

```python
#to make a data into a machine readable so that its easy for machine learning...
from sklearn import preprocessing
print(car_df["fuel"].unique())
le_fuel = preprocessing.LabelEncoder()
le_fuel.fit(car_df["fuel"])
car_df["fuel"] = le_fuel.transform(car_df["fuel"])
print(car_df["fuel"].head())
```

```python
# box plot of the selling price of the cars....
sns.boxplot(data=car_df, y="selling_price")
```

```python
#the median price based on fuel type....
plt.figure(figsize=(8,4))
sns.barplot(data=car_df, x='fuel', y="selling_price", ci=None, estimator=np.median, palette="Blues_d")
plt.title("the median price based on fuel type", fontsize=20)
```

```python
color_palette = sns.color_palette("Pastel1") + sns.color_palette("Pastel2")
labels = car_df.dtypes.value_counts().index
values = car_df.dtypes.value_counts().values
plt.pie(values, labels=labels, shadow=True, colors=color_palette, autopct="%1.1f%%")
plt.title("Pie Breakdown for Data Types")
    
white_center = plt.Circle(xy=(0,0), radius=.35, fc="white")
fig = plt.gcf()
fig.gca().add_artist(white_center)
plt.show()
```
